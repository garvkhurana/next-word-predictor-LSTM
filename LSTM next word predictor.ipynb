```python
import re
```

```python
corpus=re.sub('[^a-zA-Z0-9]'," ",corpus)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([corpus])
```

```python
words=tokenizer.texts_to_sequences([corpus])[0]
```

```python
tokenizer.index_word
```

```python
tokenizer.lower
```

```python
words
```

```python
tokenizer.word_index
```

```python
import numpy as np
x = []
y = []

for i in range(3, len(words)):
    x.append(words[i-3:i])
    y.append(words[i])

```

```python
x = np.array(x)
y = np.array(y)

print("The Length of x and y are:", len(x), len(y))
```

```python
x
```

```python
y
```

```python
total_words=len(tokenizer.word_index)+1
```

```python
import tensorflow
y=tensorflow.keras.utils.to_categorical(y,num_classes=total_words)
```

```python
y
```

```python
from tensorflow.keras.utils import pad_sequences
x=tensorflow.keras.utils.pad_sequences(x,maxlen=500,padding='pre')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=50)
```

```python
from tensorflow.keras.layers import Dense,LSTM,Embedding
from tensorflow.keras.models import Sequential
```

```python
model=Sequential()
```

```python
model.add(Embedding(total_words,10,input_length=500))
```

```python
model.add(LSTM(1000))
model.add(LSTM(1000))
model.add(Dense(1000, activation="relu"))
```

```python
model.add(Dense(total_words,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=70,validation_split=0.3,verbose=1,batch_size=62)
```
